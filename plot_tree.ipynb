```r
library(tidyverse)
library(ggtree)
```

```r
raw_tree <- read.tree("data/dengue/tree/tree_raw.nwk")
raw_tree
```

```r
tree <- read.tree("data/dengue/tree/tree.nwk")
tree
```

```r
ggtree(tree, ignore.negative.edge=TRUE) #+ scale_x_continuous(trans='log10')

options(repr.plot.width=15, repr.plot.height=14)
```

```r
ggtree(tree, ignore.negative.edge=TRUE) #+ scale_x_continuous(trans='log10')

options(repr.plot.width=15, repr.plot.height=14)
```
