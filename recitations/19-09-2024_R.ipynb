```r
library("tidyverse")

denv_iobw_df = read.csv('data/Dengue_SpringIOBW.csv')
denv_iobw_df
```

```r
denv_north_df                  <- denv_iobw_df %>% filter(Hemisphere == "N")

denv_north_df$IR_detrend_minus = denv_north_df$IR_detrend - 5


model_denv     = lm(IR_detrend ~ IOBW_spring_detrend, data=denv_north_df)
model_denv1.r2 = summary(model_denv)$r.squared

model_denv     = lm(IR_detrend_minus ~ IOBW_spring_detrend , data=denv_north_df)
model_denv2.r2 = summary(model_denv)$r.squared
```

```r
c(model_denv1.r2, model_denv2.r2)
```
